```python
from dotenv import load_dotenv
from langchain_nvidia_ai_endpoints import ChatNVIDIA
from langchain_core.messages import HumanMessage    
from langchain.agents import create_agent
from langchain_core.tools import tool
from langgraph.checkpoint.memory import InMemorySaver  

import os
load_dotenv()
```

```python
model = ChatNVIDIA(
    model="deepseek-ai/deepseek-v4-pro",
    api_key=os.getenv("NVIDIA_API_KEY"),
    verbose=True,
)
```

```python
system_prompt = """

    You are ChefBot, a Philippine-based personal chef assistant. When given an image or text of kitchen ingredients, identify all visible ingredients and then search the web for recipes using those ingredients. 
    Present 2-3 recipe options with clear instructions. For follow-up questions, answer helpfully about cooking techniques, substitutions, or any culinary topic.
    
    Create the output with this structure.

    Recipe Name: The name of the recipe 
    Ingredients: Ingredients that are present
    Instructions: Make an ordinal number out of it for order e.g. 1., 2., 3., etc....
    
"""

config = {"configurable": {"thread_id": "1"}}

ingredients = HumanMessage(content="I have ground pork, egg, molo wrapper, onion, tomato, beef, sinigang mix, magic sarap, salt, pepper, eggplant, water spinach, radish")

chef_assistant = create_agent(
    model=model,
    system_prompt=system_prompt,
    checkpointer=InMemorySaver(),
)

response = chef_assistant.invoke(
    {"messages": [ingredients]},
    config,
)

print(response['messages'][-1].content)
```
